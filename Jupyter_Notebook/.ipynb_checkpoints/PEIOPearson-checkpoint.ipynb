# Test di Pearson

```python
import os
import math
import numpy as np
from pandas import *
import pandas as pd
from datetime import datetime
%matplotlib inline 
import matplotlib.pyplot as plt
from statsmodels.distributions.empirical_distribution import ECDF
from scipy.stats import norm
from scipy.stats import genextreme,gumbel_r
from numpy import linspace
from pylab import plot,show,hist,figure,title

```

```python
def colore(c):
    colori = ["red","blue","cyan","green","black"]
    if '1h' in c:
        return colori[0]
    elif '3h' in c:
        return colori[1]
    elif '6h' in c:
        return colori[2]
    elif '12h' in c:
        return colori[3]
    elif '24h' in c:
        return colori[4]

def gumbel(x,abs_d,col):
    return np.exp(-np.exp(-(x-abs_d[col]["a"])/abs_d[col]["b"]))

def X2(data,abs_t,h,delta): 
    dt = data[h].dropna()
    #print(dt)
    lh = len(dt)
    #print("lh: ",lh)
    q = [delta*(i+1) for i in range(int(1/delta))] #definisco quantili
    #print("q: ",q)
    rv = gumbel_r(loc=abs_t.at["a",h],scale=abs_t.at["b",h])
    ecdf = ECDF(dt)
    r = ecdf(rv.ppf(q))
    #print("r: ",r)
    o0=lh*r
    #print("o0: ",o0)
    o=o0-np.append([0],np.delete(o0,-1))
    #print("o: ",o)
    e=[delta*lh for i in range(len(r))]
    #print("e: ",e)
    return ((o-e)**2/e).sum()  #valore del chi quadro 
    
def test_pearson(file, abs_m, abs_mq, abs_mle):
    abs_peio = pd.read_pickle(abs_m)  
    abs_mq_peio = pd.read_pickle(abs_mq)
    abs_mle_peio = pd.read_pickle(abs_mle) 

    #cumulo i valori in un vettore
    data = pd.read_csv(file)
    map(datetime,data['anno'])
    data=data.set_index('anno')

    a=[]
    for  h in data.columns:
        a = a + [[X2(data,abs_peio,h,0.2),X2(data,abs_mq_peio,h,0.2),X2(data,abs_mle_peio,h,0.2)]]
    v = pd.DataFrame(a,index=["1h","3h","6h","12h","24h"],columns=["M","Mq","MLE"])
    print(v)
    
    best=DataFrame([abs_mq_peio["1h"],
                abs_mle_peio["3h"],
                abs_mle_peio["6h"],
                abs_mq_peio["12h"],
                abs_mq_peio["24h"]]).T #valori pi bassi da individuare a mano 
    print(best)
    best.to_pickle("best_param") #salvo parametri migliori
    
    t_rain=np.linspace(data.min()[0],data.max()[4],100) #creo vettore con 100 punti intermedi tra il max e min di precipitazione
    gb=pd.DataFrame([gumbel(t_rain,best,"1h"),
                 gumbel(t_rain,best,"3h"),
                 gumbel(t_rain,best,"6h"),
                 gumbel(t_rain,best,"12h"),
                 gumbel(t_rain,best,"24h")]).T
    gb.index=t_rain
    gb.columns=["1h","3h","6h","12h","24"]   #rinomino colonne
    
    gb.index=t_rain
    gb.columns=["1h","3h","6h","12h","24h"]

    durate = ["1h","3h","6h","12h","24h"]
    colori = ["red","blue","cyan","green","black"]
    plt.rc('xtick', labelsize=20) 
    plt.rc('ytick', labelsize=20) 
    plt.rcParams['figure.figsize']=(10,8)
    ax = gb.plot(color = colori)
    
    for i in durate:
        datah = data[i].dropna()
        ecdfh = ECDF(datah)
        ax.plot(datah,ecdfh(datah),'o',color = colore(i), label = i)

    ax.set_title('LSPP', size=15)
    ax.set_xlabel('Altezza di precipitazione [mm]', size=15)
    ax.set_ylabel('P [H<h]', size=15)
    ax.grid(True)
test_pearson('pioggia.txt','abs_peio','abs_mq_peio','abs_mle_peio')   

    
    
```

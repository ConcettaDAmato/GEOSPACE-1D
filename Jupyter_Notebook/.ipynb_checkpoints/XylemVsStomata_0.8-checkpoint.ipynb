# Sapflow/Cavitazion versus Stomata functioning

Here we first draw some conductivity curve, i.e. the response to cavitation of some plant, that we assume as in Tyree and Sperry(1989) is not influenced by stomatal conductance at all. Than we compare it with the atmosferic demand as a function of water pontential in leaves to find the plant operational point. The sapflow is estimated by using CLM5 (Kennedy et al., 2019) function, while te argument we follow is similar to the one of Manzoni et al., 2013.  

The equation used is:
$$ 
\underbrace{K(\psi_x) |\psi_l| }_{Approximated\, sapflow} = \underbrace{g(\psi_l) (e_^* -e_a)}_{Evaporation Demand}
$$

## According to CLM5 (Kennedy et al., 2019)

\begin{equation}
K(\psi) = \left(\frac{1}{2}\right)^{\left(\frac{\psi}{\psi_{50}}\right)^b
\end{equation}

```python
#Quantities in homogeneous units, for example psi50 = 2 MPa then psi in MPa
def CLM5(psi,psi50,b):
    return 0.5**((psi/psi50)**b)
```

## $g$ vs suction

We estimate now the evaporative demand
$$
g(\psi_l) (e*e^{} -e_a)
$$
with
$$
e^\#(T) = e^*(T_{ref}) \exp\left[{-\frac{\lambda\, m_{H_2O}}{R} \left(\frac{1}{T}-\frac{1}{T_{ref}}\right)} \right] \exp{\left(\frac{\psi_l V_{H_2O}}{RT}\right)} 
$$

```python
def waterStress(psi, psi50, a,b):
    if a< 0 or a >1:
        print('Parameters a Not in Range [0,1]')
        sys.exit()
    elif b < 1:
        print('Parameters b Not in Range [1, ...]')
        sys.exit()   
    if psi < a*psi50:
        return 1
    elif psi < b*psi50:
        return 1/((a-b)*psi50)*psi - b/(a-b)
    else:
        return 0
```

```python
COLUMN_NAMES=['Psi [MPa]','G_2','G_3','G_4']
#del dff
dff = pd.DataFrame(columns=COLUMN_NAMES)
psi=np.arange(0,30.05,0.05)
dff['Psi [MPa]']=psi
alist2=[]
alist3=[]
alist4=[]
for x in psi:
    alist2.append(0.8*waterStress(x,2,0.25,5))
    alist3.append(0.8*waterStress(x,3,0.25,5))
    alist4.append(0.8*waterStress(x,4,0.25,5))
    
dff['G_2']=alist2
dff['G_3']=alist3
dff['G_4']=alist4
dff=dff.set_index('Psi [MPa]')
dff
```

```python
dff.plot()
```

```python
#The following code was created with chatGPT inputing the LateX formula!

import math

def calculate_e(T, psi_l):
    e_star_T_ref = 0.611  # Saturation water vapor pressure at T_ref [Pa]
    V_H2O = 18e-6  # Molar volume of water [m^3/mol]
    lambd = 2.25e6  # Enthalpy of vaporization of water [J/kg]
    R = 8.31446261815324  # Universal gas constant [J/(K mol)]
    m_H2O = 18.01528e-3  # Molar mass of water [kg/mol]
    T_ref = 273.15  # Reference temperature [K]

    exponent = - (lambd * m_H2O / R) * (1 / T - 1 / T_ref) + (psi_l*V_H2O) / (R * T)
    result = math.exp(math.log(e_star_T_ref) + exponent)
    return result
```

```python
calculate_e(273.15,0.103)
```

```python
psi=np.arange(0,30.05,0.05) # Various psi in MPa
T = 303.15 #leaf temperature
Ta= 203.15 #air temperature
e_a = calculate_e(Ta,0)
e_psi=[]
for x in psi:
    e_psi.append(calculate_e(T,x*10e6)) #Please observe that we transform MPa in Pa
COLUMN_NAMES = ['Psi [MPa]','e [Pa]']
edf=pd.DataFrame(columns=COLUMN_NAMES)
edf['Psi [MPa]']=psi
edf['e [Pa]']=e_psi
edf
    
    
```

```python
edf=edf.set_index('Psi [MPa]')
edf.plot()
```

## Building the second member of the equation (1)

```python
psi=np.arange(0,30.05,0.05) # Various psi in MPa
T = 303.15 #leaf temperature
Ta= 203.15 #air temperature
e_a = calculate_e(Ta,0)
E_2=[]
E_3=[]
E_4=[]
for x in psi:
    E_2.append((calculate_e(T,x*10e6)-e_a)*waterStress(x,2,0.25,5)) #Please observe that we transform MPa in Pa
    E_3.append((calculate_e(T,x*10e6)-e_a)*waterStress(x,3,0.25,5)) #Please observe that we transform MPa in Pa
    E_4.append((calculate_e(T,x*10e6)-e_a)*waterStress(x,4,0.25,5)) #Please observe that we transform MPa in Pa
COLUMN_NAMES = ['Psi [MPa]','E_2']
edff=pd.DataFrame(columns=COLUMN_NAMES)
edff['Psi [MPa]']=psi
edff['E_2']=E_2
edff['E_3']=E_3
edff['E_4']=E_4
edff
```

```python
edff=edff.set_index("Psi [MPa]")
edff.plot()
```
